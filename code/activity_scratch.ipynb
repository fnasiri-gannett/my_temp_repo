```python
# Create a table with all the non-zero campaign durations
temp=df[~np.isnan(df.cycle_duration)][['idadvertiser', 'cycle_duration']]
```

```python
# This is where the magic happens. Loop through all the ids from the last step,
# the if there is an id for which there are cycles with missing length and started 
# on the last day of December, swap it witht average
for i in range(0,df.shape[0]):
    id = df.iloc[i,5] # 5 correspondes to idadvertiser
    if id in ids:
        if ( (pd.isnull(df.iloc[i,36])) & (df.iloc[i,25]==pd.to_datetime('2017-12-31 00:00:00')) ):
            df.iloc[i,36]=temp[temp.idadvertiser==id]['cycle_duration'].values[0] # 5 correspondes to duration
```

##### We need to create a target column. Flowing what Keri did in the past, there are a number of categories:
##### - Churn: Went away and never came back
##### - Change up: increased the budget
##### - Change down: decreased the budget
##### - Idle up: came back after a period of inactivity
##### - Idle down: Not too sure what this is.

##### The problem is that Keri's table breaks everything down to months. So it's possible to track a single campaign month to month. 
##### This data is rolled up. So for the time being let's do a binarychurn column.

##### We also need to detect seasonal behavior. If it's typical for an adveriser to go MIA from time to time then it's not churm.

```python
# To check if seasonality exists, we need to bring in statsmodels.tsa.stattools.acf then pass to it 
# all those advertisers/master advertisers where there is a zero somewhere in their budgets and 
# and count the number of significant periods.
```

```python
from statsmodels.tsa.stattools import acf
```

```python
temp = df[df.campaign_budget==0]
```

```python
adv_ids=temp.idadvertiser.unique()
```

```python
df['zero_budget_seasonal']=0
```

```python
for id in adv_ids:
    temp=df[df.idadvertiser==id][['Cycle_Number', 'campaign_budget']]
    index_list=df.index[df.idadvertiser==id].tolist()
    temp.sort_values(by=['Cycle_Number'], inplace=True)
    signal_length=temp.shape[0]
    if signal_length < 2:
        continue
    else:
        my_acf, conf=acf(temp.campaign_budget, nlags= signal_length-1, alpha=0.05)
        counter=-1 # To take care if lag = 0
        for i in range(0,len(my_acf)):
            if (my_acf[i] >=0):
                if my_acf[i]>=conf[i][1]:
                    counter +=1
            else:
                if my_acf[i] <= conf[i][0]:
                    counter+=1

        if counter >=0:
            df.loc[index_list,'zero_budget_seasonal']=1
```
